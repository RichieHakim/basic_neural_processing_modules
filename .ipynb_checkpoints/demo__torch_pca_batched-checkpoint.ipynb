make synthetic latents

```python
X_dims_fake = [5000, 200000]
dtype = torch.float32

latents = [
          torch.vstack([torch.sin(torch.linspace(0, 140, X_dims_fake[0])),
                        torch.cos(torch.linspace(2,19,X_dims_fake[0])),
                        torch.linspace(0,1,X_dims_fake[0]),
                        torch.cos(torch.linspace(0,17,X_dims_fake[0])) >0]).T,
          torch.tensor(scipy.signal.savgol_filter(np.random.rand(X_dims_fake[1], 4), 15, 3, axis=0))-0.5,
        ]

for ii, latent in enumerate(latents):
    latents[ii] = latent.type(dtype)
    
```

mix latents into data

```python
data_mix = torch.einsum('ij,jk->ik', latents[0], latents[1].T).T

print('data_mix size:')
torch_helpers.tensor_sizeOnDisk(data_mix);
```

show latents and mixed data

```python
plt.figure()
plt.plot(latents[0])

plt.figure()
plt.plot(latents[1])

plt.figure()
plt.imshow(data_mix[:1000,:200], aspect='auto')
```

do PCA

```python
DEVICE = torch_helpers.set_device(use_GPU=True, verbose=False)
components, scores, singVals, EVR = decomposition.torch_pca(data_mix,
                                                               device=DEVICE,
                                                               mean_sub=True,
                                                               zscore=False,
                                                               rank=data_mix.shape[1],
                                                               return_cpu=True,
                                                               return_numpy=False)
```

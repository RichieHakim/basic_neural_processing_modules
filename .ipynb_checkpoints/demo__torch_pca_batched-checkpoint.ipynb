make synthetic latents

```python
X_dims_fake = [5000, 20077]
dtype = torch.float32

latents = [
          torch.vstack([torch.sin(torch.linspace(0, 140, X_dims_fake[0]))*2,
                        torch.cos(torch.linspace(2,19,X_dims_fake[0]))*2,
                        torch.linspace(0,1,X_dims_fake[0])*2,
                        torch.cos(torch.linspace(0,17,X_dims_fake[0])) >0]).T*1,
          torch.tensor(scipy.signal.savgol_filter(np.random.rand(X_dims_fake[1], 4), 15, 3, axis=0))-0.5,
        ]

for ii, latent in enumerate(latents):
    latents[ii] = latent.type(dtype)
    
```

mix latents into data

```python
data_mix = torch.einsum('ij,jk->ik', latents[0], latents[1].T).T

print('data_mix size:')
torch_helpers.tensor_sizeOnDisk(data_mix);
```

Incremental PCA implementation(s)

```python
import sklearn.decomposition
```

```python
X.shape
```

```python
# X = torch.as_tensor(data_mix, dtype=torch.float32, device=DEVICE)
X = np.array(data_mix, dtype=np.float32)
n_batches=8
rank = 20
```

```python
idx_batches = list(misc.make_batches(np.arange(X.shape[0]), num_batches=n_batches))
```

```python
ipca = sklearn.decomposition.IncrementalPCA(n_components=rank,
                                             whiten=False,
                                             copy=True,
                                             batch_size=1000)
```

```python
tic = time.time()
for iter_batch, idx_batch in enumerate(idx_batches):
    assert  len(idx_batch) >= ipca.n_components   ,   'RH error: batch size is smaller than rank of decomposition'
    ipca.partial_fit(X[idx_batch])
print(time.time() - tic)
```

```python
%matplotlib notebook

fig, axs = plt.subplots(5)
for ii in range(len(axs)):
    axs[ii].plot(ipca.components_[ii])
    axs[ii].plot(components[:,ii])
```
